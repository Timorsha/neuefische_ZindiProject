```python
# precip NaN -> 0
import math

def replace_nan_zero(x):
    if math.isnan(x):
        return 0.0
    else :
        return float(x)
features_zero=["precip"]
for feature in features_zero : 
    train[feature]=train[feature].apply(lambda x: [replace_nan_zero(X) for X in x])
    test[feature]=test[feature].apply(lambda x: [replace_nan_zero(X)  for X in x])
```

```python
# remove nans
def trafo(x,y):
    y_fft = []
    x_fft = []
    for i, value in enumerate(y):
        if not math.isnan(value):
            y_fft.append(value)
            x_fft.append(x[i])
    return x_fft,y_fft
```

```python
def cn(time,y,period,n):
   c = y*np.exp(-1j*2*n*np.pi*time/period)
   return c.sum()/c.size

def f(x, y, time, period, Nh):
   f = np.array([2*cn(time,y,period,i)*np.exp(1j*2*i*np.pi*x/period) for i in range(1,Nh+1)])
   return f.sum()

def fillna_fft(y_fft,x_fft,x,N_coeff):
    period = 24
    time = np.array(x_fft)
    time2 = np.array(x)
    mn = np.mean(y_fft)
    y = np.array(y_fft) - mn
    y2 = np.array([f(t,y,time,period,N_coeff).real for t in time2])
    rmse = np.sum((y2[x_fft]-y)**2)**0.5
    y2[x_fft]=y
    y2 = y2 + mn
    return y2, rmse
```

```python
def fft_n_coeff_search(n_max,y_fft,x_fft,x):
    N=1
    err_new = 0
    for n in range(n_max):
        _,err = fillna_fft(y_fft,x_fft,x,n)
        if n == 1: 
            err_new = err
        else:
            if err < err_new:
                err_new = err
                N=n
    return N
```

```python
# define nan-values via fourier transformation
import time

start = time.time()

train_final = train.copy()
features_fft=["temp","rel_humidity","wind_dir","wind_spd","atmos_press"]
x = [i for i in range(121)]
n_max = 30
for fea in features_fft:
    for i in range(len(train_final)):   
          y_raw = train.loc[i,fea]
          if np.nan in y_raw:
              x_fft,y_fft = trafo(x,y_raw)
              N_coeff = fft_n_coeff_search(n_max,y_fft,x_fft,x)
              y_new,_ = fillna_fft(y_fft,x_fft,x,N_coeff)
              train_final.at[i,fea] = y_new.tolist()
              #print(y_raw)
              #print(train_final.loc[i,fea])
              #plt.plot(x,y_new_list)
              #plt.plot(x,y_raw)

end = time.time()
running_time = end - start
print(running_time)
```
